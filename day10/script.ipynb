```python
with open("input.txt") as f:
    commands = list(map(lambda x: x.strip(), f.readlines()))
```

```python
X = 1
scoresPerRun = []

for command in commands:
    splitted_command = command.split(" ")
    scoresPerRun += [X]
    if len(splitted_command) == 2:
        scoresPerRun += [X]
        addedValue = int(splitted_command[1])
        X += addedValue

scoresPerRun
```

```python
interestingRuns = [20, 60, 100, 140, 180, 220]
result = list(map(lambda x: x*scoresPerRun[x-1], interestingRuns))
sum(result)
```

```python
currentLine = ""
lines = []
for i in range(0, 240):
    currentPixelIndex = i % 40
    currentX = scoresPerRun[i]
    print(currentPixelIndex, currentX, abs(currentX-currentPixelIndex))
    if abs(currentX-currentPixelIndex) <= 1:
        currentLine += "#"
    else:
        currentLine += "."

    if currentPixelIndex == 39:
        lines += [currentLine]
        currentLine = ""
```
