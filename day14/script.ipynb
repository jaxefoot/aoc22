```python
with open("input.txt") as f:
    lines = list(map(lambda x: x.strip(), f.readlines()))

rock = []

def getPoint(pointString):
    x,y = pointString.split(",")
    return (int(x), int(y))

def generateRock(point1, point2, rock):
    if point1[0] == point2[0]:
        yCoordinates = [point1[1], point2[1]]
        yCoordinates.sort()
        yCoordinates[1]+=1
        for y in range(*yCoordinates):
            if (point1[0], y) not in rock: rock.append((point1[0], y))
    elif point1[1] == point2[1]:
        xCoordinates = [point1[0], point2[0]]
        xCoordinates.sort()
        xCoordinates[1]+=1
        for x in range(*xCoordinates):
             if (x, point1[1]) not in rock: rock.append((x, point1[1]))

for line in lines:
    points = line.split(" -> ")
    points = list(map(lambda x: getPoint(x), points))
    for i in range(0, len(points)):
        if i<len(points)-1:
            generateRock(points[i], points[i+1], rock)
rock
```

```python
maximumRockY = max(list(map(lambda rock: rock[1], rock)))
maximumRockY
```

```python
def getFinalPlacementOfSand(startingPoint, maximumRockY, occupiedSpace):
    sand = startingPoint
    while True:
        # Task 2
        if sand[1] == maximumRockY+1:
            return sand
        if (sand[0], sand[1]+1) not in occupiedSpace:
            # Task 1
            # if sand[1]+1 > maximumRockY:
            #     return None
            sand = sand[0], sand[1]+1
        elif (sand[0]-1, sand[1]+1) not in occupiedSpace:
            sand = sand[0]-1, sand[1]+1
        elif (sand[0]+1, sand[1]+1) not in occupiedSpace:
            sand = sand[0]+1, sand[1]+1
        else:
            return sand

stopSand = False
sandCounter = 0
occupiedSpace = dict(zip(rock, [True]*len(rock)))
while (not stopSand):
    newSandPosition = getFinalPlacementOfSand((500, 0), maximumRockY, occupiedSpace)
    # Task 1
    # stopSand = newSandPosition == None
    # Task 2
    stopSand = newSandPosition == (500, 0)
    sandCounter += 1
    occupiedSpace[newSandPosition]=True
sandCounter
```
