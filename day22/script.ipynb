```python
with open("input.txt") as f:
    lines = list(map(lambda x: x.replace("\n", ""), f.readlines()))

tiles = set()
walls = set()
directions = [(1,0),(0,1),(-1,0),(0,-1)]

for i in range(0, len(lines)):
    line = [*lines[i]]
    if len(line) == 0:
        break

    column = 1
    while len(line) > 0:
        character = line.pop(0)
        if character == "#":
            walls.add((column, i+1))
        elif character == ".":
            tiles.add((column, i+1))
        column += 1
```

```python
from functools import reduce
import operator

def addTuples(*tuples):
    return tuple([sum(tup) for tup in zip(*tuples)])

def multiplyTuples(*tuples):
    return tuple([reduce(operator.mul, tup, 1)  for tup in zip(*tuples)])
```

```python


def rotate(directions, currentDirection, rotationDirection):

    currentIndex = directions.index(currentDirection)
    if rotationDirection == "L":
        return directions[currentIndex-1]
    else:
        return directions[(currentIndex+1)%len(directions)]

def getNextPosition(current, direction, tiles, walls):
    nextPotentialPosition = addTuples(current, direction)
    if nextPotentialPosition not in tiles and nextPotentialPosition not in walls:
        tempDirection = multiplyTuples(direction, (-1,-1))
        nextPotentialPosition = current
        while True:
            tempNextPotentialPosition = addTuples(nextPotentialPosition, tempDirection)
            if tempNextPotentialPosition not in tiles and tempNextPotentialPosition not in walls:
                break
            nextPotentialPosition = tempNextPotentialPosition
    if nextPotentialPosition in walls: return current
    else: return nextPotentialPosition

getNextPosition((12, 6), (1,0), tiles, walls)
```
