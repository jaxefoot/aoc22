```python
import pandas as pd
with open("input.txt") as f:
    lines = f.readlines()
    lines = list(map(lambda x: [*x.strip()], lines))
originalMap = pd.DataFrame(lines)

def getHeight(character):
    if character == "S":
        return 0
    elif character == "E":
        return ord("z")-ord("a")
    else:
        return ord(character)-ord("a")

heightMap = originalMap.apply(lambda x: x.map(lambda a: getHeight(a)))
heightMap
```

```python


class Node:
    def __init__(self, coordinates, path, height, cost, letter):
        self.coordinates = coordinates
        self.path = path
        self.height = height
        self.cost = cost
        self.letter = letter

    def __str__(self) -> str:
        return str(self.coordinates) + "path: [" + ", ".join([str(coordinate) for coordinate in self.path]) + "], cost:" + str(self.cost)+ ", height:"+ str(self.height)+ ", letter: "+ self.letter

    def __updatePath__(self, path, cost):
        self.path = path
        self.cost = cost


def checkIfPointInDataframe(point, dataframe):
    return point[0]>=0 and point[1] >= 0 and point[0] < dataframe.shape[1]and point[1] < dataframe.shape[0]


def updateNodesForAPoint(coordinates, allNodes, stack, heightMap, originalMap):
    node = allNodes[coordinates]
    aboveCoordinates = (node.coordinates[0], node.coordinates[1]-1)
    belowCoordinates = (node.coordinates[0], node.coordinates[1]+1)
    leftCoordinates = (node.coordinates[0]-1, node.coordinates[1])
    rightCoordinates = (node.coordinates[0]+1, node.coordinates[1])
    neighborPoints = [aboveCoordinates, belowCoordinates, leftCoordinates, rightCoordinates]
    neighborPoints = [point for point in neighborPoints if checkIfPointInDataframe(point, heightMap)]
    for neighborPoint in neighborPoints:
        neighborHeight = heightMap.iloc[neighborPoint[1], neighborPoint[0]]
        currentPathCost = len(node.path) + 1 if neighborHeight - node.height <= 1 else float("inf")
        if neighborPoint in allNodes.keys():
            neighborPointNode = nodes[neighborPoint]
            if currentPathCost < neighborPointNode.cost:
                neighborPointNode.__updatePath__(path = node.path+[node.coordinates], cost=currentPathCost)
                stack.append(neighborPoint)
        elif currentPathCost != float("inf"):
            neighborLetter = originalMap.iloc[neighborPoint[1], neighborPoint[0]]
            newNode = Node(coordinates=neighborPoint, path=node.path+[node.coordinates], height=neighborHeight, cost=currentPathCost, letter=neighborLetter)
            allNodes[neighborPoint] = newNode
            #print("CREATE", newNode)
            stack.append(neighborPoint)
    
def findShortestPath(startingCoordinates, goalLetter):
    nodes = {}
    nodes[startingCoordinates] = Node(coordinates=startingCoordinates, path=[], height=0, cost=0, letter="S")
    stack = [startingCoordinates]
    while len(stack)> 0:
        coordinates = stack.pop(0)
        #print("CALCULATING: ", coordinates)
        updateNodesForAPoint(coordinates, nodes, stack, heightMap=heightMap, originalMap=originalMap)
    result =min([node.cost for node in nodes.values() if node.letter == goalLetter])
    return result

#for node in nodes.values(): print(node)
```

```python
findShortestPath((0,20), "E")
findShortestPath((20,46), "a")
findShortestPath((20,46), "S")
```
