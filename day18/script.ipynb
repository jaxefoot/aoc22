```python
with open("input.txt") as f:
    lines = f.readlines()
    lines = list(map(lambda x: x.strip(), lines))
```

```python
lavaPoints = set()

xValues,yValues, zValues = [], [], []

for line in lines:
    x,y,z = list(map(lambda x: int(x), line.split(",")))
    xValues+=[x]
    yValues+=[y]
    zValues+=[z]
    lavaPoints.add((x,y,z))
maxX, maxY, maxZ= max(xValues),max(yValues),max(zValues)

openSides = 0

for point in lavaPoints:
    x,y,z = point
    if (x+1, y,z) not in lavaPoints: openSides+=1
    if (x-1, y,z) not in lavaPoints: openSides+=1
    if (x, y+1,z) not in lavaPoints: openSides+=1
    if (x, y-1,z) not in lavaPoints: openSides+=1
    if (x, y,z+1) not in lavaPoints: openSides+=1
    if (x, y,z-1) not in lavaPoints: openSides+=1

openSides
```

```python
outsidePoint = (maxX+1,maxY+1,maxZ+1)
outsidePoints = set()
insidePoints = set()
outSides = 0

def getNeighbouringPoints(point):
    x,y,z = point
    return [(x+1, y,z),
    (x-1, y,z),
    (x, y+1,z),
    (x, y-1,z),
    (x, y,z+1),
    (x, y,z-1)]

def checkIfPointIsOutsidePoint(startPoint, endpoint, blockerPoints, insidePoints, outSidePoints):
    foundPoints = set()
    stack=[startPoint]
    while len(stack) > 0:
        currentPoint = stack.pop(0)
        if currentPoint in outSidePoints:
            outSidePoints = outSidePoints.union(foundPoints)
            return True
        elif currentPoint in insidePoints:
            insidePoints = outSidePoints.union(foundPoints)
            return False
        neighbouringPoints = getNeighbouringPoints(currentPoint)
        neighbouringPoints = list(filter(lambda point: point not in foundPoints, neighbouringPoints))
        neighbouringPoints = list(filter(lambda point: point not in blockerPoints, neighbouringPoints))
        foundPoints.update(neighbouringPoints)
        stack+=neighbouringPoints
        if endpoint in foundPoints:
            outSidePoints = outSidePoints.union(foundPoints)
            return True
    insidePoints = insidePoints.union(foundPoints)
    return False

for lavaPoint in lavaPoints:
    neighbours = getNeighbouringPoints(lavaPoint)
    neighbours = list(filter(lambda point: point not in lavaPoints, neighbours))
    neighbours = list(filter(lambda point: checkIfPointIsOutsidePoint(point, outsidePoint, lavaPoints, insidePoints, outsidePoints), neighbours))
    outSides+=len(neighbours)
outSides
```
